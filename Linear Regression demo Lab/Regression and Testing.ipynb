````python
slope, intercept = np.polyfit(x, y, 1)
predicted_y = slope * x + intercept

plt.scatter(x, y, s=10, label="Data with noise")
plt.plot(x, predicted_y, label="Regression line")

plt.title("Random Data with Regression Line")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.show()

print("Slope:", slope)
print("Intercept:", intercept)
````

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

x = np.linspace(-5, 5, 500)
x = np.random.choice(x, 50)

true_y = 5 * x - 2

y_noise = np.random.normal(loc=0, scale=50, size=len(x))
y = true_y + y_noise

slope, intercept = np.polyfit(x, y, 1)
predicted_y = slope * x + intercept

sorted_indices = np.argsort(x)
x_sorted = x[sorted_indices]
true_y_sorted = true_y[sorted_indices]
predicted_y_sorted = predicted_y[sorted_indices]

plt.scatter(x, y, s=10, label="Data with strong noise")
plt.plot(x_sorted, true_y_sorted, label="True line: y = 5x - 2")
plt.plot(x_sorted, predicted_y_sorted, label="Estimated regression line")

plt.title("Strong Noise Regression Demo")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.show()

print("Estimated slope:", slope)
print("Estimated intercept:", intercept)
````

````python
mse = np.mean((y - predicted_y) ** 2)

print("Mean Squared Error:", mse)
print("Estimated slope:", slope)
print("Estimated intercept:", intercept)

assert slope > 0
print("Test passed: regression has positive trend")
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

x = np.linspace(-5, 5, 500)
x = np.random.choice(x, 50)

true_y = 5 * x - 2
noise = np.random.normal(loc=0, scale=50, size=len(x))
y = true_y + noise

manual_a = 1
manual_b = 0
manual_prediction = manual_a * x + manual_b

true_prediction = 5 * x - 2

estimated_a, estimated_b = np.polyfit(x, y, 1)
estimated_prediction = estimated_a * x + estimated_b

manual_mse = np.mean((y - manual_prediction) ** 2)
true_mse = np.mean((y - true_prediction) ** 2)
estimated_mse = np.mean((y - estimated_prediction) ** 2)

sorted_indices = np.argsort(x)
x_sorted = x[sorted_indices]

plt.scatter(x, y, s=10, label="Real data with noise")
plt.plot(x_sorted, manual_prediction[sorted_indices], label="Manual model: y = 1x + 0")
plt.plot(x_sorted, true_prediction[sorted_indices], label="True model: y = 5x - 2")
plt.plot(x_sorted, estimated_prediction[sorted_indices], label="Estimated regression")

plt.title("Model Comparison")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()

comparison = pd.DataFrame({
    "Model": [
        "Manual model",
        "True model",
        "Estimated regression"
    ],
    "a / slope": [
        manual_a,
        5,
        estimated_a
    ],
    "b / intercept": [
        manual_b,
        -2,
        estimated_b
    ],
    "MSE": [
        manual_mse,
        true_mse,
        estimated_mse
    ]
})

comparison
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

# ======================================================
# 1. Base data
# ======================================================

x = np.linspace(-5, 5, 500)
x = np.random.choice(x, 50)

true_y = 5 * x - 2

# ======================================================
# 2. Weak noise data
# ======================================================

weak_noise = np.random.normal(loc=0, scale=1, size=len(x))
y_weak = true_y + weak_noise

# ======================================================
# 3. Strong noise data
# ======================================================

strong_noise = np.random.normal(loc=0, scale=50, size=len(x))
y_strong = true_y + strong_noise

# ======================================================
# 4. Manual model
# ======================================================

manual_a = 1
manual_b = 0

manual_prediction = manual_a * x + manual_b

# ======================================================
# 5. Estimated regression models
# ======================================================

weak_a, weak_b = np.polyfit(x, y_weak, 1)
weak_prediction = weak_a * x + weak_b

strong_a, strong_b = np.polyfit(x, y_strong, 1)
strong_prediction = strong_a * x + strong_b

true_prediction = 5 * x - 2

# ======================================================
# 6. MSE calculations
# ======================================================

def mse(real, predicted):
    return np.mean((real - predicted) ** 2)

comparison = pd.DataFrame({
    "Case": [
        "Perfect data / True model",
        "Weak noise / Manual model",
        "Weak noise / True model",
        "Weak noise / Estimated regression",
        "Strong noise / Manual model",
        "Strong noise / True model",
        "Strong noise / Estimated regression"
    ],
    "Model": [
        "y = 5x - 2",
        "y = 1x + 0",
        "y = 5x - 2",
        f"y = {weak_a:.2f}x + {weak_b:.2f}",
        "y = 1x + 0",
        "y = 5x - 2",
        f"y = {strong_a:.2f}x + {strong_b:.2f}"
    ],
    "MSE": [
        mse(true_y, true_prediction),
        mse(y_weak, manual_prediction),
        mse(y_weak, true_prediction),
        mse(y_weak, weak_prediction),
        mse(y_strong, manual_prediction),
        mse(y_strong, true_prediction),
        mse(y_strong, strong_prediction)
    ]
})

# Sort x values for smooth lines
sorted_indices = np.argsort(x)
x_sorted = x[sorted_indices]

# ======================================================
# 7. Graph 1 - Perfect line
# ======================================================

plt.scatter(x, true_y, s=10, label="Perfect data")
plt.plot(x_sorted, true_prediction[sorted_indices], label="True line: y = 5x - 2")

plt.title("1. Perfect Linear Data")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()

# ======================================================
# 8. Graph 2 - Weak noise
# ======================================================

plt.scatter(x, y_weak, s=10, label="Weak noise data")
plt.plot(x_sorted, true_prediction[sorted_indices], label="True model: y = 5x - 2")
plt.plot(x_sorted, weak_prediction[sorted_indices], label="Estimated regression")

plt.title("2. Weak Noise Regression")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()

# ======================================================
# 9. Graph 3 - Strong noise
# ======================================================

plt.scatter(x, y_strong, s=10, label="Strong noise data")
plt.plot(x_sorted, true_prediction[sorted_indices], label="True model: y = 5x - 2")
plt.plot(x_sorted, strong_prediction[sorted_indices], label="Estimated regression")

plt.title("3. Strong Noise Regression")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()

# ======================================================
# 10. Graph 4 - Manual vs True vs Estimated
# ======================================================

plt.scatter(x, y_strong, s=10, label="Strong noise data")
plt.plot(x_sorted, manual_prediction[sorted_indices], label="Manual model: y = 1x + 0")
plt.plot(x_sorted, true_prediction[sorted_indices], label="True model: y = 5x - 2")
plt.plot(x_sorted, strong_prediction[sorted_indices], label="Estimated regression")

plt.title("4. Manual vs True vs Estimated Model")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()

# ======================================================
# 11. Results table
# ======================================================

comparison
````
