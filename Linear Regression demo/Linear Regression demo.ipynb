```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 50)
y = 5 * x - 2

plt.plot(x, y)
plt.grid(True)
plt.show()
```

```python
plt.scatter(x, y, s=20, label="Data with noise")

plt.title("Noisy Linear Data")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.show()
```

```python
slope, intercept = np.polyfit(x, y, 1)

print("Estimated slope:", slope)
print("Estimated intercept:", intercept)
```

```python
predicted_y = slope * x + intercept

plt.scatter(x, y, s=20, label="Data with noise")
plt.plot(x, predicted_y, label="Regression line")

plt.title("Linear Regression Fit")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.show()
```

```python
error = np.mean((y - predicted_y) ** 2)

print("Mean Squared Error:", error)

assert slope > 4 and slope < 6
assert intercept > -5 and intercept < 1

print("Tests passed")
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

x = np.linspace(-5, 5, 50)
true_y = 5 * x - 2

noise = np.random.normal(0, 4, size=x.shape)
y = true_y + noise

slope, intercept = np.polyfit(x, y, 1)
predicted_y = slope * x + intercept

plt.scatter(x, y, s=20, label="Data with noise")
plt.plot(x, true_y, label="True line: y = 5x - 2")
plt.plot(x, predicted_y, label="Estimated regression line")

plt.title("True Line vs Estimated Regression Line")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.show()

print("Estimated slope:", slope)
print("Estimated intercept:", intercept)
```

```python
error = np.mean((y - predicted_y) ** 2)

print("Mean Squared Error:", error)

assert slope > 4 and slope < 6
assert intercept > -5 and intercept < 1

print("Tests passed")
```
